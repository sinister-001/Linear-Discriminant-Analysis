```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Read the dataframe

```python
df = pd.read_csv('gender.csv', index_col=0)
df.rename(columns={'Unnamed: 1': 'Class'}, inplace=True)
df
```

### Label the traget variables

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Class'] = le.fit_transform(df['Class'])
df
```

### Divide the features and target variables

```python
X = df.drop('Class', axis = 1)
y = df['Class']
```

### Define the final required dimension count

```python
# Define the number of dimensions (d) and the desired reduced dimensionality (d')
d = X.shape[1]  # 128 dimensions
d_prime = 2  # You can change this value based on your requirements
```

### Seperate the data into train and test samples

```python
male_sample = X[0:10]
female_sample = X[400:410]
X_test = pd.concat([male_sample, female_sample], axis = 0)
male_sample = y[0:10]
female_sample = y[400:410]
y_test = pd.concat([male_sample, female_sample], axis = 0)
```

```python
X_train = X.drop(X_test.index, axis = 0).reset_index(drop = True)
y_train = y.drop(y_test.index, axis = 0).reset_index(drop = True)
```

### Compute Mean Vectors for each class

```python
X_train = X_train.to_numpy()
y_train = y_train.to_numpy()
mean_male = np.mean(X_train[y_train == 0, :], axis=0)
mean_female = np.mean(X_train[y_train == 1, :], axis=0)
```

### Within Class Scatter Matrix

```python
# Initialize the within-class scatter matrix S_W
S_W = np.zeros((d, d))  # d is the number of features (128 in your case)

# Calculate the scatter matrix for the male class
scatter_male = np.zeros((d, d))
for x in X_train[y_train == 1, :]:
    x_minus_mean = x - mean_male
    scatter_male += np.outer(x_minus_mean, x_minus_mean)

# Calculate the scatter matrix for the female class
scatter_female = np.zeros((d, d))
for x in X_train[y_train == 0, :]:
    x_minus_mean = x - mean_female
    scatter_female += np.outer(x_minus_mean, x_minus_mean)

# Sum the scatter matrices for each class to get S_W
S_W = scatter_male + scatter_female
```

```python
S_W
```

### Between Class Scatter Matrix

```python
# Calculate the overall mean vector for the entire dataset
mean_overall = np.mean(X_train, axis=0)

# Initialize the between-class scatter matrix S_B
S_B = np.zeros((d, d))

# Calculate the scatter matrix for the male class
scatter_male1 = np.zeros((d, d))
count_male = len(X_train[y_train == 1])
mean_minus_overall_male = mean_male - mean_overall
scatter_male1 = count_male * np.outer(mean_minus_overall_male, mean_minus_overall_male)

# Calculate the scatter matrix for the female class
scatter_female1 = np.zeros((d, d))
count_female = len(X_train[y_train == 0])
mean_minus_overall_female = mean_female - mean_overall
scatter_female1 = count_female * np.outer(mean_minus_overall_female, mean_minus_overall_female)

# Sum the weighted scatter matrices for each class to get S_B
S_B = scatter_male1 + scatter_female1
```

```python
S_B
```

### Eigen Values and Eigen Vectors

```python
# Calculate the eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(
    np.linalg.inv(scatter_male + scatter_female).dot(S_B)
)

# Sort the eigenvalues and eigenvectors
eigen_pairs = [(eigenvalues[i], eigenvectors[:, i]) for i in range(len(eigenvalues))]
eigen_pairs = sorted(eigen_pairs, key=lambda x: x[0], reverse=True)

# Choose the top d' eigenvectors and create the transformation matrix W
W = np.array([eigen_pair[1] for eigen_pair in eigen_pairs[:d_prime]]).T
```

### Get the final dataframe with d' dimensions

```python
# Project the data onto the reduced-dimensional space
X_train_lda = X_train.dot(W)
X_test_lda = X_test.dot(W)
```

```python
X_test_lda = X_test_lda.to_numpy()
```

### We will be using KNN Classifier for modelling

#### Euclidean Distance Function

```python
def euclidean(X_train, X_test):
    all_point_distances = []
    for i in X_test:
        point_distance = []
        for j in X_train:
            dist = np.sqrt(np.sum((j-i)**2))
            point_distance.append(dist)
        all_point_distances.append(point_distance)
    arr = np.array(all_point_distances)
    return arr
```

#### KNN Model Function

```python
def knn_classify(distance_array, y_train, k):
    array_indexes = []
    for point in distance_array:
        nearest_points = np.argsort(point)[:k]
        array_indexes.append(nearest_points)
    array_indexes = np.array(array_indexes)
    
    y_values = []
    for point in array_indexes:
        values = []
        for i in point:
            values.append(y_train[i])
        y_values.append(values)
    y_pred = []
    for point in y_values:
        y_pred.append(max(point, key = point.count))
    y_pred = np.array(y_pred)
    return y_pred
```

#### Accuracy Function

```python
def check_accuracy(y_pred, y_test):
    return np.sum(y_pred == y_test) / len(y_test)
```

```python
y_test = y_test.to_numpy()
```

### Final Evaluation

```python
eud = euclidean(X_train_lda, X_test_lda)
k = 3
y_pred = knn_classify(eud, y_train, k)
print(f"Actual Output: {y_test}\nPredicted Output: {y_pred}")
```

```python
print(f"Accuracy with k={k} : {check_accuracy(y_pred, y_test)*100}%")
```
